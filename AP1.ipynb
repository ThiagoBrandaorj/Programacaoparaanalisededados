## QUESTÃO 02 (1,0 PONTOS)

Defina uma função chamada `square_range` com 3 parâmetros (os mesmos da função `range`):

- `start`: nº inicial da sequência gerada pela função `range` (c/ valor padrão 0)
- `stop`: nº onde a sequência gerada pela função `range` para (não incluído na sequência)
- `step`: diferença entre números consecutivos na sequência (c/ valor padrão 1)

A função `square_range` deve usar *list comprehension* para criar uma lista com os quadrados dos números gerados pela função range (com os argumentos recebidos) e retorná-la.

Use a função `square_range` para criar 5 listas de acordo com as especificações abaixo e as imprima:

- Definindo valores para os 3 parâmetros.
- Definindo apenas o valor para o parâmetro `stop`.
- Definindo valores apenas para os parâmetros `start` e `stop`.
- Definindo valores apenas para os parâmetros `stop` e `step`.
- Definindo apenas um parâmetro na função.


````python
def square_range(start=0, stop=None ,step=1):
  if stop is None:
    stop = start
    start = 0
  return [x ** 2 for x in range(start, stop , step)]
print(square_range(1,15, 3))
print(square_range(stop=5))
print(square_range(start=3,stop=6))
print(square_range(stop=7,step=2))
print(square_range(start=10))
````

Se um número inválido for digitado, o programa deve ignorá-lo, mostrando uma breve mensagem de aviso, e voltando a pedir outro número. Após o final da votação, o programa deverá exibir:

- O percentual de votos de cada um destes jogadores;
- O número do jogador escolhido como o melhor jogador da partida, juntamente com o número de votos e o percentual de votos dados a ele.

Observe que os votos inválidos e o zero final não devem ser computados como votos. O resultado aparece ordenado pelo número do jogador. O programa deve fazer uso de listas e deverá executar o cálculo do percentual de cada jogador através de uma função. Esta função receberá dois parâmetros: o número de votos de um jogador e o total de votos, calculará o percentual e retornará o valor calculado.

Abaixo segue uma tela de exemplo, cuja disposição deverá ser reproduzida.

````python
print("Enquete quem foi o melhor jogador ?")
total_votos = 0
n_jogadores = [0] * 24
while True:
  n = int(input("Número do jogador (0=fim): "))
  if n > 23 or n < 0:
    print("Informe um valor entre 1 e 23 ou 0 para sair!")
    continue
  if n == 0:
    break
  n_jogadores[n] += 1
  total_votos += 1
print("Resultado da votação: ")
print(f"Foram computados {total_votos} votos.")
melhor_jog = 0
votos_melhor_jog = 0
for jog , votos in enumerate(n_jogadores):
  if votos > 0:
    percentual = (votos / total_votos) * 100
    print(f"Jogador {jog}: {votos} votos ({percentual:.1f}%)")
    if votos > votos_melhor_jog:
      melhor_jog = jog
      votos_melhor_jog = votos
print(f"O melhor jogador foi o número {melhor_jog}, com {votos_melhor_jog} votos, correspondendo a {(votos_melhor_jog / total_votos) * 100 : .1f} do total de votos")
````

## QUESTÃO 05 (1,0 PONTOS)
Faça um programa que peça os parâmetros a, b e c de uma equação do segundo grau e calcule suas raízes.

Lembrando que uma equação do segundo grau é da forma:

$a \cdot x^2 + b \cdot x + c = 0$

````python
import cmath
a = float(input("Digite o A: "))
if a == 0:
  raise Exception("'A' não pode ser zero")
b = float(input("Digite o B: "))
c = float(input("Digite o C: "))

delta = (b**2) - (4 * a * c)
if delta < 0:
  raiz1 = (-b + cmath.sqrt(delta)) / (2*a)
  raiz2 = (-b - cmath.sqrt(delta)) / (2*a)
  print(f"Raiz1: {raiz1} , Raiz2: {raiz2}")
elif delta == 0:
  raiz = (-b) / (2*a)
  print(f"Raiz: {raiz}")
else:
  raiz1 = (- b + (delta**0.5)) / (2*a)
  raiz2 = (- b - (delta**0.5)) / (2*a)
  print(f"Raiz 1: {raiz1} , Raiz 2: {raiz2}")
````

## QUESTÃO 06 (1,0 PONTOS)

Faça um programa que calcule o número dado pela sequência de Fibonacci para um número inserido pelo usuário. Seu programa deve funcionar de forma recursiva e não armazenar valores intermediários de forma alguma!

Por definição, a sequência de Fibonacci é definida por:

$F_n = F_{n-1} + F_{n-2}$

Com $F_1 = F_2 = 1$, e $F_0 = 0$.

Qual o maior Fibonacci que sua rotina calcula em 10s de processamento?

````python
import time
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1 or n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

def main():
  n = 0
  max_fib = 0
  cronometo_iniciado = time.time()
  while True:
     fib = fibonacci(n)
     if fib > max_fib:
      max_fib = fib
     tempo_passado = time.time() - cronometo_iniciado
     if tempo_passado >= 10:
         break
     n += 1
  print(f"O maior fibonacci calculado em 10 segundos é: {max_fib}")
if __name__ == "__main__":
  main()
````

## QUESTÃO 07 (1,0 PONTOS)

Faça um novo programa que calcule o número dado pela sequência de Fibonacci para um número inserido pelo usuário. Porém agora, você deverá armazenar os resultados intermediários do cálculo para poder calcular números de Fibonacci maiores. Qual o maior Fibonacci que sua rotina calcula em 10s de processamento?

````python
import time
def fibonacci(n, memo={}):
    if n in memo:
        return memo[n]
    if n == 0:
        return 0
    elif n == 1 or n == 2:
        return 1
    else:
        result = fibonacci(n-1, memo) + fibonacci(n-2, memo)
        memo[n] = result
        return result

n = int(input("Insira um número para calcular na sequência de Fibonacci: "))
cronometro_iniciado = time.time()
fibonacci_number = fibonacci(n)
tempo_passado = time.time() - cronometro_iniciado
print("O número de Fibonacci correspondente é:", fibonacci_number)
print("Tempo de processamento:", tempo_passado, "segundos")
````
